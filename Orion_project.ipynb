```python
# Import libraries and modules
import plotly.express as px
# Import matplotlib
import matplotlib.pyplot as plt
# Import 3D visualization library
from mpl_toolkits.mplot3d import Axes3D
```

```python
# View given x,y,z coordinates
x = [-0.41, 0.57, 0.07, 0.00, -0.29, -0.32,-0.50,-0.23, -0.23]
y = [4.12, 7.71, 2.36, 9.10, 13.35, 8.13, 7.19, 13.25,13.43]
z = [2.06, 0.84, 1.56, 2.07, 2.36, 1.72, 0.66, 1.25,1.38]
```

```python
# Create a figure 2D
fig = plt.figure()

# Add your subplot
ax = fig.add_subplot(1,1,1)
# Create a scatter plot
plt.scatter(x,y)
plt.show()
```

```python
# Create a figure 3D
fig_3d = plt.figure()
# Add your subplot
ax = fig_3d.add_subplot(1, 1, 1, projection="3d")

# Create a scatter plot
plt.scatter(x,y,z)
plt.show()
```

```python
# Challenge - Create a figure 2D with white points and black background (as sky)
fig = plt.figure()

# Add your subplot
ax = fig.add_subplot(1,1,1)
ax.set_facecolor((0.0, 0.0, 0.0))
# Create a scatter plot (white points)
ax.scatter(x,y,color=(1.0, 1.0, 1.0))
plt.show()
```

```python
# Challenge - Create a figure 3D with blue points (as constallation)
fig_3d = plt.figure()
# Add your subplot
ax = fig_3d.add_subplot(1, 1, 1, projection="3d")
#black background
ax.set_facecolor((0.0, 0.0, 0.0))
ax.grid(False)
ax.w_xaxis.pane.fill = False
ax.w_yaxis.pane.fill = False
ax.w_zaxis.pane.fill = False

# Create a scatter plot
ax.scatter(x,y,z, color=(0.46,0.68,1.0))
plt.show()
```
