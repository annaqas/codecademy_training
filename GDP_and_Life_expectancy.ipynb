# **Import Libraries**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

# **Explore data**

```python
data = pd.read_csv('/content/sample_data/all_data.csv')
print(data.head())
```

```python
print(data.info())
```

```python
data.Country.unique()
```

```python
data.Year.unique()
```

```python
le_mean = data['Life expectancy at birth (years)'].mean()
le_std = data['Life expectancy at birth (years)'].std()
print('The life expectancy of all data are: ' + str(le_mean) + ' with standard deviation of the ' + str(le_std))
```

```python
gdp_mean = data.GDP.mean()
gdp_std = data.GDP.std()
print('The gross domestic product of all country on this database are: ' + str(gdp_mean) + ' with std of the ' + str(gdp_std))
```

# **Data analysis and visualization**

```python
# What the difference of life expectancy means between countries?

plt.gcf().set_size_inches(12, 8)
sns.boxplot(x = "Country", y = "Life expectancy at birth (years)", palette = "pastel", data = data)
plt.title('Mean of the life expectancy at birth (years) by country')
plt.xticks(rotation=0)
plt.show()
```

```python
#What is the behaviour of the life expectancy over the years 2000-2015, by country?

plt.gcf().set_size_inches(12, 8)

le_chile = data["Life expectancy at birth (years)"][data.Country == "Chile"]
ano_chile = data.Year[data.Country == "Chile"]
plt.plot(ano_chile, le_chile)

le_china = data["Life expectancy at birth (years)"][data.Country == "China"]
ano_china = data.Year[data.Country == "China"]
plt.plot(ano_china, le_china)

le_ger = data["Life expectancy at birth (years)"][data.Country == "Germany"]
ano_ger = data.Year[data.Country == "Germany"]
plt.plot(ano_ger, le_ger)

le_mex = data["Life expectancy at birth (years)"][data.Country == "Mexico"]
ano_mex = data.Year[data.Country == "Mexico"]
plt.plot(ano_mex, le_mex)

le_usa = data["Life expectancy at birth (years)"][data.Country == "United States of America"]
ano_usa = data.Year[data.Country == "United States of America"]
plt.plot(ano_usa, le_usa)

le_zim = data["Life expectancy at birth (years)"][data.Country == "Zimbabwe"]
ano_zim = data.Year[data.Country == "Zimbabwe"]
plt.plot(ano_zim, le_zim)

plt.title('Mean of the life expectancy at birth (years) by country over time')
plt.legend(data.Country.unique())
plt.xlabel("Years")
plt.ylabel("Life expectancy at birth (years)")

plt.show()
```

```python
# What the difference of GDP means between countries, between 2000-2015?

plt.gcf().set_size_inches(12, 8)
sns.boxplot(x='Country', y='GDP', data=data, palette='Accent')
plt.title('Mean of the GDP by country')
plt.xticks(rotation=0)
plt.show()
```

```python
#What is the behaviour of the GDP over the years 2000-2015, by country?

plt.gcf().set_size_inches(12, 8)

gdp_chile = data.GDP[data.Country == "Chile"]
ano_chile = data.Year[data.Country == "Chile"]
plt.plot(ano_chile, gdp_chile)

gdp_china = data.GDP[data.Country == "China"]
ano_china = data.Year[data.Country == "China"]
plt.plot(ano_china, gdp_china)

gdp_ger = data.GDP[data.Country == "Germany"]
ano_ger = data.Year[data.Country == "Germany"]
plt.plot(ano_ger, gdp_ger)

gdp_mex = data.GDP[data.Country == "Mexico"]
ano_mex = data.Year[data.Country == "Mexico"]
plt.plot(ano_mex, gdp_mex)

gdp_usa = data.GDP[data.Country == "United States of America"]
ano_usa = data.Year[data.Country == "United States of America"]
plt.plot(ano_usa, gdp_usa)

gdp_zim = data.GDP[data.Country == "Zimbabwe"]
ano_zim = data.Year[data.Country == "Zimbabwe"]
plt.plot(ano_zim, gdp_zim)

plt.title('Mean of the GDP by country over time')
plt.legend(data.Country.unique())
plt.xlabel("Years")
plt.ylabel("Gross Domestic Product in USD")

plt.show()
```

```python
# limits of LE by country

max_chile = np.max(data["Life expectancy at birth (years)"][data.Country == "Chile"])
min_chile = np.min(data["Life expectancy at birth (years)"][data.Country == "Chile"])
print(min_chile, max_chile)

max_china = np.max(data["Life expectancy at birth (years)"][data.Country == "China"])
min_china = np.min(data["Life expectancy at birth (years)"][data.Country == "China"])
print(min_china, max_china)

max_ger = np.max(data["Life expectancy at birth (years)"][data.Country == "Germany"])
min_ger = np.min(data["Life expectancy at birth (years)"][data.Country == "Germany"])
print(min_ger, max_ger)

max_mex = np.max(data["Life expectancy at birth (years)"][data.Country == "Mexico"])
min_mex = np.min(data["Life expectancy at birth (years)"][data.Country == "Mexico"])
print(min_mex, max_mex)

max_usa = np.max(data["Life expectancy at birth (years)"][data.Country == "United States of America"])
min_usa = np.min(data["Life expectancy at birth (years)"][data.Country == "United States of America"])
print(min_usa, max_usa)

max_zim = np.max(data["Life expectancy at birth (years)"][data.Country == "Zimbabwe"])
min_zim = np.min(data["Life expectancy at birth (years)"][data.Country == "Zimbabwe"])
print(min_zim, max_zim)
```
