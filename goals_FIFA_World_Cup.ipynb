```python
from matplotlib import pyplot as plt
import pandas as pd
import seaborn as sns

df=pd.read_csv('/content/sample_data/WorldCupMatches.csv')

df['Total Goals'] = df['Home Team Goals'] + df['Away Team Goals']
df.dropna(inplace=True)
df['Year_'] = df['Year'].astype('int') 
print(df.info())

sns.set_style('whitegrid')
f, ax = plt.subplots(figsize=(14,10))
ax = sns.barplot(data=df, x='Year_', y='Total Goals')
sns.set_context("poster", font_scale = .5, rc={"grid.linewidth": 0.6})
ax.set_title('Average Number Of Goals Scored In FIFA World Cup By Year', fontsize=20)
plt.show()
plt.clf()

sns.set_style('whitegrid')
f, ax2 = plt.subplots(figsize=(14,10))
ax2 = sns.boxplot(data=df, x='Year_', y='Home Team Goals', palette='Spectral')
sns.set_context("notebook", font_scale = 1.25)
ax2.set_title('FIFA World Cup goals scored by year', fontsize=20)
plt.show()
```
